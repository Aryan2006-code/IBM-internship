```python
# data type
#list
my_list =[1,2,3,4,5]

#method
my_list.append(6) #append() method adds an item to the end of the list.
print(my_list) #output: [1, 2, 3, 4, 5, 6]

my_list.insert(0,10) #insert() method inserts an item at a given position.  
print(my_list) #output: [10, 1, 2, 3, 4, 5, 6]  

my_list.remove(3) #remove() method removes the first occurrence of a specified value.
print(my_list) #output: [10, 1, 2, 4, 5, 6]

my_list.pop() #pop() method removes the item at the given position in the list, and returns it. If no index is specified, pop() removes and returns the last item in the list.
print(my_list) #output: [10, 1, 2, 4, 5]


#properties
#mutable and ordered
my_list[0] = 100 #lists are mutable, so we can change the value of an item in the list.
print(my_list) #output: [100, 1, 2, 4,

#heterogenerous & dublicates allowed
zen = [1,2,3,45,"hello", True, 2, 3] #lists can contain items of different data types and can have duplicate values.
print(zen) #output: [1, 2, 3, 45, 'hello', True, 2, 3]
```

```python
# class 

class Car:
    #attributes "variables"
    
    def __init__(self, color, model, year):
        self.color = color
        self.model = model
        self.year = year

    #methods "function"
    def start(self):
        print(f"The {self.color} {self.model} is starting.")    


    def stop(self):
        print(f"The {self.color} {self.model} is stopping.")

    def play(self , song):
        print(f"The {self.color} {self.model} is playing {song}.")    

bmw = Car("blue", "BMW X5", 2020)
```

```python
bmw.start() #output: The blue BMW X5 is starting.
bmw.play("Shape of You") #output: The blue BMW X5 is playing Shape of You.  
bmw.stop() #output: The blue BMW X5 is stopping.
```

```python
#create a class box , having some attributes  and placing methids (stack, push , pop ,append)
class Box :

    def __init__(self , size , shape , color , volume):
        self.size = size
        self.shape = shape
        self.color = color
        self.volume = volume

    def stack(self):
        print(f"Rohan is stacking a box of color: {self.color}, shape: {self.shape} , size : {self.size} and volume : {self.volume}")

    def push(self):
        print(f"The {self.color} {self.shape} box is pushing.")

    def pop(self):
        print(f"The {self.color} {self.shape} box is popping.") 

md = Box("large", "square", "red", 100)

md.stack()
md.push()
md.pop()
```
